```python
def list_files_in_directory(directory):
    
        # Check if the directory exists
        if not os.path.exists(directory):
            print(f"Directory '{directory}' not found.")
        
        # List files in the directory
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        
        return files
    

# Example usage
directory_path = "Data Base/stationYerly"  
file_list = list_files_in_directory(directory_path)




print("List of files in the directory:")
print(file_list)
```

# Prepares one tabla from all the stations in the folder

!define it as a function and call it also in line 13

```python
merged_df =   pd.read_csv("Processed Data/station yearly processed/"+file_list[0])
merged_df = merged_df.rename(columns={ merged_df.columns[1]: f'{file_list[0][:-4]}'})
for f in range(1,len(file_list)-1):

    
    
    df1 = pd.read_csv("Processed Data/station yearly processed/"+file_list[f])
    column_name1 = str(file_list[f][:-4]) 
    df1 = df1.rename(columns={df1.columns[1]: column_name1})
    
    merged_df = pd.merge(df1,  merged_df
                     , on='date',how='outer'  )
pd.set_option('display.max_rows', None)
merged_df = merged_df.sort_values(by='date').reset_index(drop=True)
merged_df   
```

## Now we will find stations that have missing data

```python
nan_counts = merged_df.isna().sum()

# Display the NaN counts for each column
print(nan_counts)
```

!

```python
nan_counts_beginning = merged_df.apply(lambda col: col.isna().astype(int).groupby(col.notna().astype(int).cumsum()).cumsum().max())

# Display the counts of consecutive NaNs at the beginning for each column
print(nan_counts_beginning)
```

```python
#Now we will find the number of missing data (from the beginning of the measurement)
nan_counts_after_cuont = nan_counts-nan_counts_beginning

# you can Display the NaN counts for each column
#print(nan_counts_after_cuont)
```

```python
cell_counts = merged_df.count()
#print(cell_counts)
```

```python
prosant_of_mising_data = 100*(nan_counts_after_cuont)/(cell_counts+nan_counts_after_cuont)

print(prosant_of_mising_data)
```

### Trims the data to produce a one line to balance the start of the data measurement as much as possible

!

```python
# Assuming your DataFrame is named df_filled
merged_df = merged_df.iloc[6:]  # This will keep rows starting from the 7th row (index 6)

# Reset the index if needed
merged_df.reset_index(drop=True, inplace=True)
```

### replace NaN with average

```python
# Replace NaN with the column-wise average for numeric columns only
numeric_cols = merged_df.select_dtypes(include='number').columns
df_filled = merged_df.copy()
df_filled[numeric_cols] = merged_df[numeric_cols].apply(lambda col: col.fillna(col.mean().astype(int)), axis=0)
```
