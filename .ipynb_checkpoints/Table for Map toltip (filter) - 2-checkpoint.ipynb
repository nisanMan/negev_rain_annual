```python
import pandas as pd
import os
import numpy as np
```

###### A list of the names in the source information folder "Appears in several places:DB preparing (English, deleting)"

```python
def list_files_in_directory(directory):
    
        # Check if the directory exists
        if not os.path.exists(directory):
            print(f"Directory '{directory}' not found.")
        
        # List files in the directory
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        
        return files
    


directory_path = "Processed Data/station yearly processed"  
file_list = list_files_in_directory(directory_path)




print("List of files in the directory:")

print(file_list)

```

# Prepares one tabla from all the stations in the folder

### A function that creates a DF from a list in files (name : "name.csv" )

```python
def filter_by_opening_date(file_list):
    merged_df =   pd.read_csv("Processed Data/station yearly processed/"+file_list[0])
    merged_df = merged_df.rename(columns={ merged_df.columns[1]: f'{file_list[0][:-4]}'})
    for f in range(1,len(file_list)):

    
    
        df1 = pd.read_csv("Processed Data/station yearly processed/"+file_list[f],encoding='utf-8')
        column_name1 = str(file_list[f][:-4]) 
        df1 = df1.rename(columns={df1.columns[1]: column_name1})
    
        merged_df = pd.merge(df1,  merged_df
                     , on='date',how='outer'  )
    #pd.set_option('display.max_rows', None)
    merged_df = merged_df.sort_values(by='date').reset_index(drop=True)
    return merged_df    
```

```python
merged_df = filter_by_opening_date(file_list)

pd.set_option('display.max_columns', None)

merged_df.head()
```

## Now we will find stations that have missing data

```python
nan_counts = merged_df.isna().sum()



# Display the NaN counts for each column
nan_counts
```

```python
nan_counts_beginning = merged_df.apply(pd.Series.first_valid_index)
```

```python
#Now we will find the number of missing data (from the beginning of the measurement)
nan_counts_after_cuont = nan_counts-nan_counts_beginning

# you can Display the NaN counts for each column
print(nan_counts_after_cuont)
```

### total NaN

```python
cell_counts = merged_df.count()
#print(cell_counts)
```

### Percentage calculation of missing data 

```python
prosant_of_mising_data = 100*(nan_counts_after_cuont)/(cell_counts+nan_counts_after_cuont)

print(prosant_of_mising_data)
print(len(prosant_of_mising_data))
```

### We will manually delete everything above 10% of deficiencies

```python
indices_to_remove = []
for f in range(len(prosant_of_mising_data)):
    if prosant_of_mising_data[f] > 10:
        indices_to_remove.append(len(prosant_of_mising_data)-f-1)

print(indices_to_remove)
#HAR HORESHA,MAMSHIT KURNUB,,ROAD 40 - NAHAL ZIN  CROSSING ,HAR HATIRA ASCEN  bor nitzana do not exist
#indax:9,10,24,30
indices_to_remove = [ 23, 15, 13, 4, 3]
print(indices_to_remove)
```

```python

file_list = [file for i, file in enumerate(file_list) if i not in indices_to_remove]



file_list
```
