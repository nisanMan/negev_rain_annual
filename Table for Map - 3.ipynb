```python
import pandas as pd
import numpy as np
```

### A little analysis

```python
df_filled = pd.read_csv("Processed Data/Good stations for toltip map.csv") 
df_filled = df_filled.describe().T
df_filled = df_filled.reset_index()
df_filled = df_filled.drop(df_filled.index[0])

df_filled =df_filled.drop(["count"],axis = 1)

df_filled = df_filled.rename(columns={df_filled.columns[0]: "station name"})
df_filled.head()
```

```python
df_filled.to_csv("Processed Data/stations analysis.csv")
```

## Development of the tabla for the map

```python
df2 = pd.read_csv('Processed Data/meta stations rain.csv')
df2.head()
```

```python
merged_df = df2.merge(df_filled, how='inner', on='station name')
merged_df.head()
```

```python
column_indices_to_drop = [0, 1, 4, 5, 6,10, 11, 12, 14, 16, 17, 18]
merged_df = merged_df.drop(merged_df.columns[column_indices_to_drop], axis=1)
merged_df.head()
```
